Genero i vettori $Z_j$ tramite una distribuzione gaussiana normale. $$Z_1, \ldots, Z_k \qquad f: \mathbb{R}^d \mapsto \mathbb{R}^k$$

```python
from scipy.stats import norm
from functools import reduce
import random
import math

def distance(x, y): # || x - y || 
    return math.sqrt(sum([ math.pow(x[i] - y[i], 2) for i in range(len(x)) ]))
```

```python
d = 1000
k = 50

Z = [ list(norm.rvs(size=d)) for j in range(k) ]

def f(x):
    def _f(j, x):
        return sum([ Z[j][i] * x[i] for i in range(d) ])
    return [ _f(i, x) / math.sqrt(k) for i in range(k) ]
```

Ora genero un insieme $S \subset \mathbb{R}^d$ di cardinalit√† $n$ e verifico la norma dei vettori, considerati a due a due, cambi a meno di una costante moltiplicativa dopo averli mappati in $\mathbb{R}^k$.

```python
n = range(100)

S = { i: [ random.randrange(10) for a in range(d) ] for i in n }
S_k = { i: f(S[i]) for i in n }

%time { (i, j): distance(S[i], S[j]) for i in n for j in n }
```
